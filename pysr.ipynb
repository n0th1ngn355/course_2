```python
import sympy
import numpy as np
from matplotlib import pyplot as plt
from pysr import PySRRegressor
from sklearn.model_selection import train_test_split
```

```python
# np.random.seed(0)
X = np.linspace(-10, 10, 100)
# y = np.array([i*i for i in X])
f = lambda x: x*np.sin(5*x)
y = f(X)
X = X.reshape(-1, 1)


# from collections import defaultdict

# points = defaultdict
# X, y = [], []
# with open("output.txt", "r") as f:
#     for i in f.readlines():
#         a, b = i.split(',')
#         if a == 'end':
#             break
#         X.append(int(a))
#         y.append(int(b))

# X = np.array(X).reshape(-1,1)
# y = np.array(y)
# y = max(y)-y
plt.plot(X, y)
```

```python
Xt = X[45:110]
yt = y[45:110]
plt.plot(X,y)
plt.plot(Xt, yt)
```

```python
default_pysr_params = dict(
    populations=30,
    model_selection="best",
)
```

```python
model = PySRRegressor(
    niterations=10,
    binary_operators=["+", "*", "-", "/"],
    unary_operators=["cos", "exp", "sin", "tan", "abs", "log"],
    **default_pysr_params,
)

model.fit(X, y)
```
